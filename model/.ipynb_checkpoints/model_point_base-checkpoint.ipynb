### Visualize the input data 

```python
def plotMap(fileName):
    mapName = 'map_patches/'+fileName
    segName = 'seg_patches/'+fileName
    labelName = 'labels/'+fileName

    map_img = mpimg.imread(filePath+mapName)
    seg_img = mpimg.imread(filePath+segName)
    label_img = mpimg.imread(filePath+labelName)
    
    plt.rcParams["figure.figsize"] = (10,6)
    plt.subplot(1,3,1)
    plt.title("map")
    plt.imshow(map_img)

    plt.subplot(1,3,2)
    plt.title("legend")
    plt.imshow(label_img) 
    
    plt.subplot(1,3,3)
    plt.title("segmentation")
    plt.spy(seg_img, markersize=5)
 
    plt.show()
    
plotMap('UT_Eureka_249207_1954_24000_geo_mosaic_3_pt_305.png')
```

```python
def blackcolorSegmentation(fileName):
    map_img = cv2.imread(fileName)
    
    kChannel = 255 - np.max(map_img, axis=2) # Calculate channel K
    binaryThresh = 120
    _, binaryImage = cv2.threshold(kChannel, binaryThresh, 255, cv2.THRESH_BINARY)
    
#     plt.subplot(1,2,1)
#     plt.title("map_img")
#     plt.imshow(map_img)

#     plt.subplot(1,2,2)
#     plt.title("Filtered img")
#     plt.imshow(binaryImage) 
#     plt.show()    

    return binaryImage

# binaryImage = blackcolorSegmentation(filePath+'labels/'+'UT_Eureka_249207_1954_24000_geo_mosaic_3_pt_305.png')
```

```python
def detectPoint(fileName):
    mapName = filePath+'map_patches/'+fileName
    labelName = filePath+'labels/'+fileName
    segName = filePath+'seg_patches/'+fileName

    map_binaryImage = blackcolorSegmentation(mapName)
    label_binaryImage = blackcolorSegmentation(labelName)
    label_binaryImage = cv2.resize(label_binaryImage, (20,20))
    res = cv2.matchTemplate(map_binaryImage, label_binaryImage, cv2.TM_CCORR_NORMED)

    map_img = cv2.imread(mapName)
    plt.subplot(1,2,1)
    plt.title("map_img")
    plt.imshow(map_img)

    lab_img = cv2.imread(labelName)
    plt.subplot(1,2,2)
    plt.title("legend img")
    plt.imshow(lab_img) 
    plt.show()

    threshold = 0.5
    res[res <= threshold] = 0

    seg_img = mpimg.imread(segName)

    plt.rcParams["figure.figsize"] = (10,4)
    plt.subplot(1,4,1)
    plt.title("filtered map")
    plt.imshow(map_binaryImage)

    plt.subplot(1,4,2)
    plt.title("filtered legend")
    plt.imshow(label_binaryImage)
    
    plt.subplot(1,4,3)
    plt.title("true seg_img")
    plt.spy(seg_img, markersize=5)

    plt.subplot(1,4,4)
    plt.title("predicted seg_img")
    plt.spy(res, markersize=5)
    plt.show()

detectPoint('UT_Eureka_249207_1954_24000_geo_mosaic_3_pt_305.png')
```

```python
filenames = os.listdir('/home/shared/DARPA/patched_category/point/training/labels') 
filenames[0]
```

```python
for fileName in random.sample(filenames, 4):
    detectPoint(fileName)
```

```python
def plotFourier(fileName):
    segName = 'seg_patches/'+fileName
    seg_img = mpimg.imread(filePath+segName)
    
    dft = cv2.dft(seg_img, flags = cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

    plt.rcParams["figure.figsize"] = (8,3)
    plt.subplot(1,2,1)
    plt.title("seg_img")
    plt.spy(seg_img, markersize=5)

    plt.subplot(1,2,2)
    plt.title("Fourier Transform")
    plt.imshow(magnitude_spectrum) 
    plt.show()
    
plotFourier('UT_Eureka_249207_1954_24000_geo_mosaic_3_pt_305.png')
```
