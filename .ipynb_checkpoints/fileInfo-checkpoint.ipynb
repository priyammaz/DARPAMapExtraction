```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from itertools import chain
from PIL import Image
Image.MAX_IMAGE_PIXELS = 933120000

import json
import glob
import pickle
```

```python
filePath = '/home/shared/DARPA/training/'
jsonFiles = glob.glob(filePath+'*.json')
```

```python
# create a dataFrame to summarize all the training files' information
nFiles = len(jsonFiles)
state = [''] * nFiles
county = [''] * nFiles
nlabels = [0] * nFiles
labelsList = [[]] * nFiles

for i, jsonFile in enumerate(jsonFiles):
    state[i], county[i] = jsonFile.split('.')[0].split('/')[-1].split('_')[0:2]
    
    with open(jsonFile, 'r') as f:
        jsonData = json.load(f)
    nlabels[i] = len(jsonData['shapes'])
    labelsList[i] = [x['label'] for x in jsonData['shapes']]
    
#     for j in range(len(jsonData['shapes'])):
#         point_coord = jsonData['shapes'][j]['points']
#         if point_coord[0][0] >= point_coord[1][0] or point_coord[0][1] >= point_coord[1][1]:
#             print("Coordinate right is less than left:", state[i], county[i])
#             print(point_coord)
            
#         flatten_list = list(chain.from_iterable(point_coord))   
#         if (len(flatten_list)!=4):
#             print("Point Coordinates number is wrong:", state[i], county[i])
#             print(point_coord)
```

```python
fileInfoDF = pd.DataFrame({'state':state,'county':county, 'nlabels':nlabels,'labelsList':labelsList})
fileInfoDF
```

```python
 # row other info
fileInfoDF.iloc[fileInfoDF['nlabels'].idxmax()]  # row's index that has the most number of labels
```

```python
ptList = []
lineList = []
polyList = []

for key, labelList in fileInfoDF.labelsList.items():
    for label in labelList:
        if label.split('_')[-1]=='poly':
            polyList.append(label)
        elif label.split('_')[-1]=='line':
            lineList.append(label)
        elif label.split('_')[-1]=='pt':
            ptList.append(label)
        else:
            print("A label that is not poly, line, pt found at: ", key)
```

```python
len(ptList), len(lineList), len(polyList) # mostly are polygon legend
```

```python
len(set(ptList)), len(set(lineList)), len(set(polyList)) # ~ half legend share the same legend name
```

```python
# top K most frequent type of pt in the ptList
from collections import Counter
occurence_count = Counter(ptList)
res=occurence_count.most_common(10)
print (res)
```

```python
occurence_count = Counter(lineList)
res=occurence_count.most_common(10)
print (res)
```

```python
occurence_count = Counter(polyList)
res=occurence_count.most_common(10)
print (res)
```

### Categories of legend

1. Pure color
2. color with text
3. color with text and pattern & Pure pattern  ->(decompose to backgrounp color, text, pattern (dot, line), pattern color). 

4. symbol for point: ['1_pt', '2_pt', '3_pt', '4_pt', '5_pt', 'flathe_pt', 'vent_pt', 'cone_pt', 'Foliatio_pt', 'folia_inclin_pt']
5. symbol for line: ['fault_line', 'mass_line', 'quarry_line', 'contact_line', 'vertical_bedding_pt', 'horiz_bedding_pt', 'overturn_bedding_pt', 'inclined_bedding_pt', 'flowline_line', 'crater_line', 'fault_thrust_line', 'fault_normal_line', 'channel_line', 'skylight_line']

```python
legendPath = '/home/shared/DARPA/legend_label/'
legendFiles = glob.glob(legendPath+'*.png')
legendFiles[0].split('.')[0].split('_')[-2:]
```

```python
pt_3_list = []
for legendFile in legendFiles:
    if legendFile.split('.')[0].split('_')[-2:] == ['3','pt']:
        pt_3_list.append(mpimg.imread(legendFile))
        
plt.figure(figsize=(20,10))
columns = 5
for i, image in enumerate(pt_3_list):
    plt.subplot(int(len(pt_3_list) / columns) + 1, columns, i + 1)
    plt.imshow(image)    
```

```python
pt_4_list = []
for legendFile in legendFiles:
    if legendFile.split('.')[0].split('_')[-2:] == ['4','pt']:
        pt_4_list.append(mpimg.imread(legendFile))
        
plt.figure(figsize=(20,10))
columns = 5
for i, image in enumerate(pt_4_list):
    plt.subplot(int(len(pt_4_list) / columns) + 1, columns, i + 1)
    plt.imshow(image)  
```

```python
normal_fault_line_list = []
for legendFile in legendFiles:
    if legendFile.split('.')[0].split('_')[-3:] == ['normal','fault','line']:
        normal_fault_line_list.append(mpimg.imread(legendFile))
        
plt.figure(figsize=(20,10))
columns = 5
for i, image in enumerate(normal_fault_line_list):
    plt.subplot(int(len(normal_fault_line_list) / columns) + 1, columns, i + 1)
    plt.imshow(image)  
```

```python
thrust_fault_line = []
for legendFile in legendFiles:
    if legendFile.split('.')[0].split('_')[-3:] == ['thrust','fault','line']:
        thrust_fault_line.append(mpimg.imread(legendFile))
        
plt.figure(figsize=(20,10))
columns = 5
for i, image in enumerate(thrust_fault_line):
    plt.subplot(int(len(thrust_fault_line) / columns) + 1, columns, i + 1)
    plt.imshow(image)  
```
