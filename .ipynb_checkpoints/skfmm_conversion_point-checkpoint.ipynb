```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import skfmm
import matplotlib.image as mpimg
import glob
import pickle
import cv2
import imageio
```

```python
filePath = '/home/shared/DARPA/all_patched_data/training/point/seg_patches/'
savedPath = '/home/shared/DARPA/all_patched_data/training/point/seg_patches_converted/'
FileList = glob.glob(filePath+'*.png')
FileList[0].split('.')[0].split('/')[-1]
```

```python
def fmm_conversion(label):
    """
    label: 2d numpy array, [256,256], sparse only a few 1's, most are 0's
    """
    label = np.where(label, 0, -1) + 0.5
    sd = skfmm.distance(label, dx = 1)
    # Normalised [0,1]
    sd_norm = (sd - np.min(sd))/np.ptp(sd)
    return sd_norm
```

```python
filename = FileList[0]
array = mpimg.imread(filename)
sd_norm = fmm_conversion(array)
subname = filename.split('.')[0].split('/')[-1]
imageio.imwrite(savedPath+subname+'.png', (sd_norm*255).astype(np.uint8))
```

```python
# only need to run once

for filename in FileList:
    array = mpimg.imread(filename)
    sd_norm = fmm_conversion(array)
    subname = filename.split('.')[0].split('/')[-1]
    plt.imsave(savedPath+subname+'.png', sd_norm)
    imageio.imwrite(savedPath+subname+'.png', (sd_norm*255).astype(np.uint8))
```
