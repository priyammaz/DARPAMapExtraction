```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import skfmm
import matplotlib.image as mpimg
import glob
import pickle
import cv2
```

```python
filePath = '/home/shared/DARPA/patched_category/line/training/seg_patches/'
savedPath = '/home/shared/DARPA/patched_category/line/training/seg_patches_converted/'
FileList = glob.glob(filePath+'*.png')
FileList[0].split('.')[0].split('/')[-1]
```

```python
firstfile = mpimg.imread(FileList[0])
# plt.spy(firstfile,markersize=5)
# plt.show()
firstfile.shape
```

```python
 def fmm_conversion(label):
    """
    label: 2d numpy array, [256,256], sparse only a few 1's, most are 0's
    """
    label = np.where(label, 0, -1) + 0.5
    sd = skfmm.distance(label, dx = 1)
    # Normalised [0,1]
    sd_norm = (sd - np.min(sd))/np.ptp(sd)
    return sd_norm
```

```python
sd_norm = fmm_conversion(firstfile)
```

```python
plt.rcParams["figure.figsize"] = (6,4)
plt.subplot(1,2,1)
plt.spy(firstfile,markersize=5)
plt.subplot(1,2,2)
plt.imshow(sd_norm)
plt.show()
```

```python
secondfile = mpimg.imread(FileList[1])
sd_norm = fmm_conversion(secondfile)
#plt.imshow(sd_norm)
sd_cut = np.where(sd_norm>0.98, 1, 0)
plt.imshow(sd_cut)
```

```python
# only need to run once

for filename in FileList:
    array = mpimg.imread(filename)
    sd_norm = fmm_conversion(array)
    subname = filename.split('.')[0].split('/')[-1]
    plt.imsave(savedPath+subname+'.png', sd_norm)
```
